```python
# 로그인하기
from selenium import webdriver

driver = webdriver.Chrome("chromedriver")
driver.implicitly_wait(3)              # 드라이버가 초기화 될때 까지 3초 대기
driver.get('https://www.danawa.com/')  # 크롬 웹브라우저를 새로 연다

# 로그인 스크립트 소스를  찾아냄
login = driver.find_element_by_css_selector('li.my_page_service > a')
login.click()
driver.implicitly_wait(3)

# my_id = "----본인 아이디 입력하세요----"
# my_pw = "----본인 패스워드 입력하세요----"

my_id = "내 아이디 넣기 "
my_pw = "내 비밀번호 넣기 "

# ID와 Password를 입력하여 클릭
# <input type="text" id="danawa-member-login-input-id" maxlength="40" name="id" class="int" placeholder="아이디/인증 이메일" value="" style="z-index: 8;">
driver.find_element_by_id('danawa-member-login-input-id').send_keys(my_id)
driver.implicitly_wait(2)

# <input type="password" maxlength="20" id="danawa-member-login-input-pwd" name="password" class="int" placeholder="비밀번호" style="z-index: 8;">
driver.find_element_by_name('password').send_keys(my_pw)
driver.implicitly_wait(2)

# <button class="btn_login" id="danawa-member-login-loginButton" onclick="return false;">로그인</button>
driver.find_element_by_css_selector('button.btn_login').click()
driver.implicitly_wait(2)

# 관심 상품 링크 클릭
# <li class="interest_goods_service">
# 			<a href="http://www.danawa.com/myPage/wishList.php" class="">
# 				<span class="my_serv_tit">관심상품</span>
# 				<span class="number_has">4</span>			</a>
# </li>
wishlist = driver.find_element_by_css_selector('li.interest_goods_service > a' ).click()
driver.implicitly_wait(2)

html_src = driver.page_source # 현재 페이지의 소스
print(html_src)

# input()

driver.close()
```

```python
# 관심 상품 리스트로 출력
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_src,'html.parser')

# <table class="tbl wish_tbl">
# 		<caption>
# 			<strong>관심상품1 목록</strong>
# 			<p>관심상품1 목록으로 이미지, 상품명, 최저가 등의 정보를 제공</p>
# 		</caption>
# 		<colgroup>
# 			<col class="check">
# 			<col class="img">
# 			<col class="info">
# 			<col class="lowest">
# 		</colgroup>
# 		<thead class="th_blind">
# 			<tr><th scope="col" class="check">선택</th>
# 			<th scope="col" class="img">상품 이미지</th>
# 			<th scope="col" class="info">상품 정보</th>
# 			<th scope="col" class="lowest">최저가 정보</th>
# 			</tr>
# 		</thead>
# 		<tbody>
# 					<tr>

wish_table = soup.select('table[class="tbl wish_tbl"]')[0]
wish_items = wish_table.select('tbody tr')
# print(len(wish_items)) # 4개

for item in wish_items: # 4회 반복
    title = item.find('div',{'class':'tit'}).text
    print(title)
    
    spec_list = item.find('dd',{'class':'spec_list'}).text
    print(spec_list)
    
    lowest = item.find('td',{'class':'lowest'})
    print(lowest.find('dt').text)  # '최저가'
    
    price = item.find('span',{'class':'price'}).text
    print(price)
    
    # link
    link = item.find('a',{'class':'link'}).get('href')
    print(link)    
    
#     sub_info = item.find('div',{'class':'sub_info clear'}).text
#     print(sub_info)
    
    print('-'*80)
```
